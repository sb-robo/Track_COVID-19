### Import Dependencies

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv('../Dataset/COVID-data.csv')

# We will consider cases for INDIA only
df = df[df.location == 'India'].reset_index()
df.head()
```

```python
df.columns
```

```python
drop_col = ['index', 'iso_code', 'continent', 'location', 'reproduction_rate',
            'stringency_index', 'population','population_density', 'gdp_per_capita', 
            'extreme_poverty', 'cardiovasc_death_rate', 'diabetes_prevalence', 
            'female_smokers', 'male_smokers', 'handwashing_facilities', 'hospital_beds_per_thousand',
            'life_expectancy', 'human_development_index']

df = df.drop(drop_col, axis=1)
df.head()
```

```python
#save dataset
df.to_csv('../Dataset/India.csv',index=False)
```
